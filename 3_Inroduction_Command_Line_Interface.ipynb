## 2. Basic commands


```python
#check present working directory
!pwd
```

```python
#list files in current directory
!ls
```

```python
#create empty folder
!mkdir test_folder
```

```python
#move to other directory
!cd test_folder
```

```python
#create empty file
!touch test_file.txt
```

```python
#open text file
!more test_file.txt
```

```python
#add text to the file
!echo "I am here" >> test_file.txt
```

## 5. Writing shell script
* Shell script is useful when you need to run long and many commands.

```python
#try make your first shell script

!echo -e "mkdir test1\nmkdir test2\nmkdir test3\ntouch test1.txt\nmv test1.txt test1\necho 'Success'" > name.sh
```

```python
!more name.sh
```

```python
#you can use any text editor to make/edit shell script e.g. nano
nano name.sh
```

```python
#run shell script
!bash name.sh
```
